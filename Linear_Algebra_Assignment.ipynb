## 1.1 Create a two-dimensional vector and plot it on a graph

## 1.2 Create a three-dimensional vector and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

threeD_vector = [.1, .9, .5]


# passing the vectors into an array structure
vectors = np.array([[0, 0, 0, .1, .9, .5]])

# basically, this creates the matrix from the array
# it returns tuples, horizontally, of the array
X, Y, Z, U, V, W = zip(*vectors)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# quiver: Plot a 3D field of arrows
ax.quiver(X, Y, Z, U, V, W, length=1)


ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

##  1.7 Find $e \times f$

\begin{align}
\vec{e} = \begin{bmatrix} 5 \\ 7 \\ 2 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\4 \\ 6 \end{bmatrix}

```
# cross product = vector multiplication

# result is a vector perpendicular to both e and f

e = [5, 7, 2]
f = [3, 4, 6]

np.cross(e, f)
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
# double pipes refer to the norm or magnitude of a vector

# we can use the pythagorean theorem to find the equation for the norm

g = [1, 1, 1, 8]
h = [3, 3, 3, 3]

print(np.linalg.norm(g))
print(np.linalg.norm(h))

print("g is longer")
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12, 11],
              [ 7, 10]])

J = np.array([[ 0,  1,  2],
              [ 7, 10,  4],
              [ 3,  2,  0]])
```

```
# determinant 

print(np.linalg.det(H))
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
print(np.linalg.inv(H))
print(np.linalg.inv(J))
```

```
# scatterplot 

import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = np.random.random((100, 4))

# slicing the data for two dimensions
x = data[:,0]
y = data[:,1]
points = data[:,2:4]

# color is the length of each vector in `points`
color = np.sqrt((points**2).sum(axis = 1))/np.sqrt(2.0)
# This retrieves the jet colormap and maps each of the float values 
# (between 0 and 1) in the array color to a 3-tuple RGB value.
rgb = plt.get_cmap('jet')(color)

# scatter plot
ax.scatter(x, y, color = rgb)
plt.show()
```
